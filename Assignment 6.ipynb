```python
import numpy as np
import matplotlib.pyplot as plt

L = 1 
k = 1 
n = 100 
x = np.linspace(0, L, n)
M_in = 1
M_out = 0 

# Calculate the mass transfer rate at each point in the tube
dMdt = np.zeros(n)
dMdt[0] = M_in
dMdt[-1] = M_out

# Solve the mass transfer equation
for i in range(1, n-1):
    dMdt[i] = (dMdt[i-1] - 2*k*dMdt[i] + dMdt[i+1]) / (2*k*L/n)

plt.plot(x, dMdt)
plt.xlabel('x (m)')
plt.ylabel('dMdt (kg/m²s)')
plt.title('Mass Transfer Rate along the Tube')
plt.show()
```
